# 4. Create a Binomial Distribution

Look at the help function for
> stats.binom.pmf

What does it do?

The function is used to model probabilistic experiments which follows binomial distribution. Binomial distribution is used to model experiments that have only one of two outcomes, success or failure. 

It takes:
* n: the total number of trials or times the experiement will carried out.
* k: a list of integers from 0 to n, inclusive
* p: the probability that the outcome of a single experiment will be a success. The value of p is between 0 to 1, inclusive.

Return value:
* a list with the same number of values and sequence as k. The return value is the probability mass function for the values in k.

````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''
    
    xbar = sum(X)/len(Y)
    ybar = sum(Y)/len(Y)
    n=len(X) #since len(X) = len(Y), we can use one of them.
    
    numer = sum([xi*yi for xi, yi in zip(X,Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) -n * xbar ** 2

    m = numer/denum #slope
    b = ybar - m*xbar #intercept
    
    return m, b

# solution
m, b = best_fit(X, Y)

print("The slope is: %1.2f" % m)
print("The intercept is: %1.2f" % b)

#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
y_predict = [i*m + b for i in X]

plt.scatter(X,Y)
plt.plot(X, y_predict, label="Best fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
# ANSWER
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, h
# ANSWER
#m, se = ???, ???
#h = ???
#return m, m-h, m+h

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h
data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]

print(mean_confidence_interval(data))

results = stats.ttest_1samp(data,170,alternative='greater')
````
