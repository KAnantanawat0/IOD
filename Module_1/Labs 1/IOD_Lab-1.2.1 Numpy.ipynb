# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib

# get matplot to show charts inline
%matplotlib inline

# If you were not using %matplotlib inline, you might need to use plt.show() 
# to display the plot in a separate window.
````

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

````python
# Finding the index of the value 1.55
np.where(population == 1.55)[0]
````

````python
population[5]
````

````python
# Finding the first index of the value 1.55
np.where(population == 1.55)[0][0]
````

````python
population[5]
````

````python
np.min(height)
np.min(weight)
````

**8. Create histograms of height and weight**

````python
help(plt.hist)
````

````python
#Plt histogram for height and weight
num_bin=15
plt.hist(weight, bins=num_bin, color='darkorchid', histtype='step')
plt.title("Weight Histogram")
plt.xlabel("Weight")
plt.ylabel("count")
plt.show()
````

````python
#Plt histogram for height and weight
num_bin=15
plt.hist(height, bins=num_bin, color='firebrick', histtype='step')
plt.title("Height Histogram")
plt.xlabel("Height")
plt.ylabel("count")
plt.show()
````

**9. Create separate boxplots of height and weight**

````python
help(plt.boxplot)
````

````python
plt.boxplot(height, showfliers=False, patch_artist=True, medianprops=dict(color = "darkgreen", linewidth=1.5), boxprops = dict(facecolor = "limegreen"))
plt.title("Boxplot - Height")
plt.xticks([1],["Height"])
plt.show()
````

````python
plt.boxplot(weight, showfliers=False, patch_artist=True, medianprops=dict(color = "crimson", linewidth=1.5), boxprops = dict(facecolor = "coral"))
plt.title("Boxplot - Weight")
plt.xticks([1],["Weight"])
plt.show()
````

**10. Create a scatterplot of height and weight**

````python
help(plt.scatter)
````

````python
plt.scatter(height, weight, c='mediumvioletred', edgecolors='midnightblue', label="Points")
plt.title('Weight VS Height', )
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(loc='upper right')
plt.show()
````
