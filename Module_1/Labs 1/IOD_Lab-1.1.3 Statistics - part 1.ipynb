# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency_USA = [15, 32, 12, 7]
frequency_APJ = [10, 20, 25, 2]
fig1, ax1 = plt.subplots(nrows=1, ncols=2)
fig1.tight_layout(h_pad=5, w_pad=10) ### NOTE TO KAY!!! Need to make this graph looks better!
explode = (0, 0, 0.1, 0)  # exercise: modify this to "explode" only the 3rd slice (i.e. 'Data Scientists')

ax1[0].pie(frequency_USA, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1[1].pie(frequency_APJ, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1[1].axis('equal')
ax1[0].set_title('USA')
ax1[1].set_title('APJ')
plt.show()
```

Note: Autopct — If you want to label the wedges with their numeric value in a pie chart, then use autopct parameter. This parameter allows us to display the percent value using string formatting. Say for example, the percent value calculated for a segment is 34.678666 and if you want to display the percent value rounded to 1 decimal place then autopct parameter should be assigned the format string ‘%1.1f’ then the wedge will be labelled with the numeric value 34.6. If you want to add a percent sign (%) to the label then use two percent signs(%%) in the format string so that the special character ‘%’ escapes itself.

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

# Create a bar plot of performance for each object
#plt.xticks(???, ???)
plt.bar(y_pos, performance)
plt.xticks(y_pos,objects)
# Use ylabel and title to label the y axis and provide a title
plt.title('Frequency of language usage observed')
plt.ylabel('Performance')
plt.show()
```
